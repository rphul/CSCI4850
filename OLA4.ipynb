<h2> Problem 1 </h2>

```python
import pandas
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
from sympy import *
init_printing(use_latex=True)
from IPython.display import display
from sklearn.decomposition import PCA
import keras
```

```python
data = np.array(pandas.read_table("https://www.cs.mtsu.edu/~jphillips/courses/CSCI4850-5850/public/WDBC.txt",
delim_whitespace=True,
header=None))
```

```python
print(data.shape)
```

```python
X = data[:,0:30] 
Y = data[:,30] 

display(X.shape)
display(Y.shape)
display(Matrix(np.unique(Y)).T)
```

```python
def mean_center(x):
    return x - np.mean(x)
Xcentered = np.apply_along_axis(mean_center,0,X)
```

```python
U,S,V = np.linalg.svd(Xcentered,full_matrices=True)
plt.plot(100.0*S/np.sum(S))
plt.ylabel('% Var')
plt.xlabel('Singular Value')
plt.show()
```

```python
100.0*(S[0]+S[1])/np.sum(S)
```

```python
D = np.zeros([X.shape[0],X.shape[1]])
np.fill_diagonal(D,S)
Xrotated = np.dot(U,D)
PCs = Xrotated[:,0:2]
PCs.shape
```

```python
plt.scatter(PCs[:,0],PCs[:,1],
color=[['red','green'][i] for i in Y.astype(int)])
plt.xlabel("PC1")
plt.ylabel("PC2")
plt.show()
```

<h4> It shows two distinct categories, but there isn't a completely clear line between the two. </h4>

<h3> Problem 2 </h3>

```python
data2 = np.array(pandas.read_table("https://www.cs.mtsu.edu/~jphillips/courses/CSCI4850-5850/public/cancer-data.txt",
delim_whitespace=True,
header=None))
```

```python
print(data2.shape)
```

```python
X = data2[:,0:9] 
Y = data2[:,9] 

display(X.shape)
display(Y.shape)
display(Matrix(np.unique(Y)).T)
```

```python
def mean_center(x):
    return x - np.mean(x)
Xcentered = np.apply_along_axis(mean_center,0,X)
```

```python
U,S,V = np.linalg.svd(Xcentered,full_matrices=True)
plt.plot(100.0*S/np.sum(S))
plt.ylabel('% Var')
plt.xlabel('Singular Value')
plt.show()
```

```python
100.0*(S[0]+S[1])/np.sum(S)
```

```python
D = np.zeros([X.shape[0],X.shape[1]])
np.fill_diagonal(D,S)
Xrotated = np.dot(U,D)
PCs = Xrotated[:,0:2]
PCs.shape
```

```python
plt.scatter(PCs[:,0],PCs[:,1],
color=[['red','green','blue','yellow','purple','pink'][i] for i in Y.astype(int)])
plt.legend()
plt.xlabel("PC1")
plt.ylabel("PC2")
plt.show()
```

<h4> The distinction between the categories isn't clear, but the red, purple, and pink are separated more from the others. Some cancers, although different, might share some characteristics with each other. The yellow, green, and blue seem like they have a lot of shared features. </h4>
