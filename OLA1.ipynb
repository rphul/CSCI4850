```python
x = np.linspace(-5,5,10)
y = 1/ (1 + pow(np.e,-x))
```

```python
plt.plot(x,y)
plt.title('$y={1}/{1+e^{-x}}$')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

Problem 3

```python
x = symbols('x')
```

```python
y = 1/(1+pow(np.e,-x))
```

```python
dydx = y.diff(x)
```

```python
dydx
```

```python
x_vals = np.linspace(-5,5,10)
y_vals = np.array([dydx.evalf(subs=dict(x=x_val)) for x_val in x_vals])
plt.plot(x_vals,y_vals)
plt.title('dy/dx')
plt.xlabel('x')
plt.ylabel('dy/dx')
plt.show()
```
